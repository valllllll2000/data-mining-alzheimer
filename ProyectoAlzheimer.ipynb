# Información sobre el conjunto de datos

- **Grupo** --> Clase
- **Edad** --> Edad
- **EDUC** --> Años de Educación
- **SES** --> Estado Socioeconómico / 1-5
- **MMSE** --> Examen del Estado Mental Mínimo
- **CDR** --> Clasificación Clínica de Demencia
- **eTIV** --> Volumen Intracraneal Total Estimado
- **nWBV** --> Volumen Total del Cerebro Normalizado
- **ASF** --> Factor de Escalamiento del Atlas

## 1. MMSE (Examen del Estado Mental Mínimo)
**Descripción:** Una herramienta de cribado ampliamente utilizada para evaluar la función cognitiva. Evalúa varios dominios cognitivos, incluyendo memoria, orientación, atención y lenguaje.  
**Puntuación:** La puntuación total varía de 0 a 30, con puntuaciones más bajas que indican un mayor deterioro cognitivo (por ejemplo, una puntuación inferior a 24 sugiere posible demencia).

## 2. CDR (Clasificación Clínica de Demencia)
**Descripción:** Una escala utilizada para evaluar la gravedad de la demencia basada en entrevistas y evaluaciones clínicas. Evalúa las habilidades cognitivas y funcionales en diferentes dominios.  
**Puntuación:** Varía de 0 (sin demencia) a 3 (demencia severa), proporcionando una clara clasificación del estado cognitivo del individuo.

## 3. eTIV (Volumen Intracraneal Total Estimado)
**Descripción:** Una medida del volumen total dentro del cráneo, que incluye el cerebro, el líquido cefalorraquídeo y otras estructuras. Se utiliza a menudo en estudios de neuroimagen.  
**Propósito:** Ayuda a contextualizar los hallazgos en la imagenología cerebral al tener en cuenta las diferencias individuales en el tamaño del cráneo al evaluar los volúmenes cerebrales.  
**Rango Típico:** El eTIV se mide a menudo en centímetros cúbicos (cm³) y generalmente se encuentra dentro del rango de 1,200 a 1,800 cm³ para adultos.  
- **Hombres:** Valores típicamente más altos, promediando alrededor de 1,400 a 1,600 cm³.  
- **Mujeres:** Valores generalmente más bajos, promediando alrededor de 1,100 a 1,400 cm³.  
**Nota:** Pueden ocurrir variaciones individuales basadas en la edad, el sexo y la demografía de la población.

## 4. nWBV (Volumen Total del Cerebro Normalizado)
**Descripción:** Una medida calculada que representa el volumen total del cerebro normalizado por el volumen intracraneal total (eTIV). Ayuda a evaluar la atrofia cerebral.  
**Propósito:** Se utiliza para comparar el volumen cerebral entre individuos mientras se controla las variaciones en el tamaño de la cabeza, haciéndolo útil en estudios de enfermedades neurodegenerativas.  
**Rango Típico:** El nWBV se expresa generalmente como un porcentaje del volumen intracraneal total. Normalmente varía del 75% al 95% del eTIV.  
Un nWBV normal podría estar alrededor del 80% al 90%, indicando un volumen cerebral saludable en relación con el volumen intracraneal total.  
Valores más bajos pueden indicar atrofia cerebral, que es común en condiciones como la enfermedad de Alzheimer.

## 5. ASF (Factor de Escalamiento del Atlas)
**Descripción:** Un factor utilizado para escalar datos de neuroimagen a un atlas estándar, permitiendo comparaciones entre sujetos. Tiene en cuenta la variabilidad individual en el tamaño y la forma del cerebro.  
**Propósito:** Ayuda a alinear y comparar imágenes cerebrales de diferentes individuos, facilitando evaluaciones más precisas en entornos de investigación y clínicos.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import plot_tree
import pandas as pd

# Cargar el archivo CSV desde la URL
url = 'https://raw.githubusercontent.com/valllllll2000/data-mining-alzheimer/refs/heads/main/alzheimer.csv'
df = pd.read_csv(url)
display(df)
# Preprocess the data (handle missing values, convert categorical features)
df = df.drop('SES', axis = 1)
df = df.drop('ASF', axis = 1)
df['MMSE'] = df['MMSE'].fillna(df['MMSE'].median())
df['Group'] = df['Group'].replace('Converted', 'Nondemented')
display(df)

df = pd.get_dummies(df, columns=['Group'], drop_first=True)
df = pd.get_dummies(df, columns=['M/F'], drop_first=True)

display(df)
X = df.drop(['Group_Nondemented'], axis=1)
y = df['Group_Nondemented']

# Create the decision tree classifier with pruning parameters
tree = DecisionTreeClassifier()  # Adjust ccp_alpha for pruning strength

# Train the model
tree.fit(X, y)

plt.figure(figsize=(20, 20))  # Adjust figure size as needed
plot_tree(tree, feature_names=X.columns, class_names=['Demented', 'Non Demented'], filled=True)
plt.show()

```

Various correlation models are used to analyze relationships between different variables, such as cognitive performance, brain structure, and biochemical markers. Here are some common types of correlation models utilized in this field:

### 1. **Pearson Correlation Coefficient**
- **Description**: Measures the linear correlation between two continuous variables.
- **Range**: Values range from -1 (perfect negative correlation) to +1 (perfect positive correlation). A value of 0 indicates no correlation.
- **Usage**: Often used to examine relationships between cognitive scores (e.g., MMSE) and neuroimaging metrics (e.g., brain volume).

### 2. **Spearman Rank Correlation**
- **Description**: Assesses the strength and direction of the association between two ranked variables.
- **Range**: Similar to Pearson, ranging from -1 to +1.
- **Usage**: Useful when the data do not meet the assumptions required for Pearson's correlation, such as non-normal distribution or ordinal data.

### 3. **Kendall's Tau**
- **Description**: A non-parametric measure of correlation that assesses the strength of association between two variables by considering the ranks of the data.
- **Range**: Values range from -1 to +1, similar to Pearson and Spearman.
- **Usage**: Often used in smaller datasets or when dealing with ordinal data.

### 4. **Partial Correlation**
- **Description**: Measures the strength of association between two variables while controlling for the effect of one or more additional variables.
- **Usage**: Helps to determine the direct relationship between two variables without the influence of confounding factors, such as age or education level.

### 5. **Multiple Regression Analysis**
- **Description**: A statistical technique that models the relationship between a dependent variable and one or more independent variables.
- **Usage**: Used to analyze how multiple factors (e.g., age, sex, education, biomarkers) simultaneously affect cognitive scores or neuroimaging outcomes.

### 6. **Structural Equation Modeling (SEM)**
- **Description**: A comprehensive statistical approach that allows for the analysis of complex relationships between observed and latent variables.
- **Usage**: Useful in evaluating theoretical models of cognitive decline in Alzheimer's and testing hypotheses about the relationships among variables.

### 7. **Multivariate Analysis of Variance (MANOVA)**
- **Description**: A statistical test used to compare means across multiple groups while considering multiple dependent variables.
- **Usage**: Useful for examining differences in cognitive performance across different stages of Alzheimer's disease.

```python
# prompt: create a pearsons correlation with all the numerical data, writing the correlation values displayed inside the matrix

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
import seaborn as sns

df = pd.read_csv('alzheimer.csv')
df = df.dropna()
demented_df = df[df['Group'] == 'Demented']

# Select only numerical columns for correlation
numerical_df = demented_df.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr(method='pearson')

# Display the correlation matrix with values inside the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Pearson Correlation Matrix')
plt.show()

# High	0.7 to 1.0	Strong relationship between the variables
# Medium	0.3 to 0.7	Moderate relationship between the variables
# Low/Inexistant	0.0 to 0.3	Weak or no linear relationship between the variables

# Positive Correlation (0 < r ≤ 1): As one variable increases, the other tends to increase as well. A perfect positive correlation is +1.
# Negative Correlation (-1 ≤ r < 0): As one variable increases, the other tends to decrease. A perfect negative correlation is -1.
# No Correlation (r = 0): There is no linear relationship between the variables.

```
